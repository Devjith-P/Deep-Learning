```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

```python
df = pd.read_csv("/home/devjith/Luminar/Machine Learning/files/lung_cancer_examples.csv")
```

```python
df
```

```python
df = df.drop(columns=['Name','Surname'])
```

```python
df
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit(x_train)
x_train = scalar.fit_transform(x_train)
x_test = scalar.fit_transform(x_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann = Sequential()
ann.add(Dense(units=4,activation='relu'))

ann.add(Dense(units=10,activation='relu'))
ann.add(Dense(units=6,activation='relu'))

ann.add(Dense(units=1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
x_train.shape
```

```python
ann.fit(x_train,y_train,epochs=100,batch_size=12) 
```

```python
ann.evaluate(x_test,y_test)
```

```python
ann.summary()
```

```python
y_pred = ann.predict(scalar.transform([[42,1,3,4]]))

if y_pred>0.5:
    print("Positive for Lung Cancer")
else:
    print("Negative for Lung Cancer")
```

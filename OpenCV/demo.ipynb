```python
gray_image = cv2.imread('/home/devjith/Luminar/Deep Learning/files/baby3.jpeg',cv2.IMREAD_GRAYSCALE)
gray_image.shape
```

```python
cv2.imshow('Gray baby',gray_image)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
# cv2.imwrite('baby_gray.png',gray_image)
```

```python
ret,binary_image = cv2.threshold(gray_image,90,255,cv2.THRESH_BINARY)
cv2.imshow('Binary',binary_image)
cv2.waitKey(3000)
cv2.destroyAllWindows()
ret
```

```python
ret,binary_image = cv2.threshold(gray_image,90,255,cv2.THRESH_OTSU)
cv2.imshow('Binary',binary_image)
cv2.waitKey(3000)
cv2.destroyAllWindows()
ret
```

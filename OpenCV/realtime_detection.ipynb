```python
import cv2
```

```python

model_face = cv2.CascadeClassifier('/home/devjith/Luminar/Deep Learning/Face_Detection/haarcascade_frontalface_default.xml')
model_eye = cv2.CascadeClassifier('/home/devjith/Luminar/Deep Learning/Face_Detection/haarcascade_eye.xml')
```

```python
video = cv2.VideoCapture('/home/devjith/Luminar/Deep Learning/files/video.mp4')
while True:
    success,img = video.read()
    if success:
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        
        face = model_face.detectMultiScale(gray,minNeighbors=10)

        for (x,y,w,h) in face:
            cv2.rectangle(img,(x,y),(x+w,y+h),color=(255,0,0),thickness=2)
            #cv2.putText(img,'Devjith',(x,y-10),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=1.2,color=(255,0,0),thickness=1)
        


        cv2.imshow('Video',img)


        if cv2.waitKey(1) & 0XFF == ord('q'):
            break
    else:
        print("Video Finished")
        break
    


video.release()
cv2.destroyAllWindows()
```

```python
import cv2
import time
```

```python
model_eye =  cv2.CascadeClassifier('/home/devjith/Luminar/Deep Learning/Face_Detection/haarcascade_eye.xml')
```

```python
video = cv2.VideoCapture(0)

while True:
    success,img = video.read()

    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    eye = model_eye.detectMultiScale(gray,minNeighbors=20)

    for (x,y,w,h) in eye:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
    
    cv2.imshow("EYE",img)

    if cv2.waitKey(1) & 0XFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```

```python
video = cv2.VideoCapture('/home/devjith/Luminar/Deep Learning/files/video.mp4')
prev_frame_time = 0
new_frame_time = 0
while True:
    success,img = video.read()

    if success:

        new_frame_time = time.time()
        fps = 1 / (new_frame_time - prev_frame_time)
        prev_frame_time = new_frame_time
        fps_text = f"FPS: {int(fps)}"

        
        cv2.putText(img, fps_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

        
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

        eye = model_eye.detectMultiScale(gray,minNeighbors=15)

        for (x,y,w,h) in eye:
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)

        cv2.imshow('Eye',img)

        if cv2.waitKey(1) & 0XFF == ord('q'):
            break
    else:
        print("Video Ended")
        break
video.release()
cv2.destroyAllWindows()
```

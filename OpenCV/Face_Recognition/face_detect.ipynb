```python
import face_recognition as fr
import os, pandas, cv2
import numpy as np
```

```python
path = '/home/devjith/Luminar/Deep Learning/files/Face_images_celeb'
```

```python
face_list = os.listdir(path)
face_list
```

```python
imgs = []
classnames = []

for i in face_list:
    imgpath = os.path.join(path,i)
    cur_img = cv2.imread(imgpath)
    imgs.append(cur_img)
    classnames.append(i.split('.')[0])
classnames
```

```python
def face_encoding(images):
    encoding = []
    for img in images:
        img= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_in_frame = fr.face_locations(img)
        print(face_in_frame)
        encoding.append(fr.face_encodings(img,face_in_frame)[0])
    return encoding
        
```

```python
encoding_known_face = face_encoding(imgs)
```

```python
video = cv2.VideoCapture(0)

while True:
    suc, img = video.read()

    img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

    face_in_frame = fr.face_locations(img2)
    

    encoded_face = fr.face_encodings(img2,face_in_frame)

    for face_loc, enc in zip(face_in_frame,encoded_face):
        matches = fr.compare_faces(encoding_known_face,enc)
        distance = fr.face_distance(encoding_known_face,enc)
        match_index = np.argmin(distance)
        if matches[match_index] == True:
            name = classnames[match_index]
            print(name)
            y1,x2,y2,x1 = face_loc
            cv2.rectangle(img,(x1,y1),(x2,y2),color=(0,0,255),thickness=2)
            cv2.putText(img,f"{name.upper()}",(x1,y1-15),cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0),2)
        else:
            y1,x2,y2,x1 = face_loc
            cv2.rectangle(img,(x1,y1),(x2,y2),color=(0,0,255),thickness=2)
            cv2.putText(img,f"Unknown",(x1,y1-15),cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0),2)


    cv2.imshow("Window",img)

    if cv2.waitKey(1) & 0xff == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```

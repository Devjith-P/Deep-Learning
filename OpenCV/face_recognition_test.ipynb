```python
import cv2
import os
```

```python
model = cv2.CascadeClassifier('/home/devjith/Luminar/Deep Learning/Face_Detection/haarcascade_frontalface_default.xml')
```

```python
video = cv2.VideoCapture(0)
path = '/home/devjith/Luminar/Deep Learning/files/faces//'

if not os.path.exists(path):
    os.makedirs(path)
    
for i in range(1000):
    suc , img = video.read()

    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    face = model.detectMultiScale(gray,minNeighbors=10)
    if len(face)>0:
        for (x,y,w,h) in face:
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
            cv2.imwrite(f"{path}/joyel{i}.jpg",img[y:y+h,x:x+w])
            
    cv2.imshow('Video',img[y:y+h,x:x+w])


    



    if cv2.waitKey(1) & 0XFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```

```python
faces = os.listdir('/home/devjith/Luminar/Deep Learning/files/faces')
dir = '/home/devjith/Luminar/Deep Learning/files/faces'
```

```python
x = []
y = []
```

```python
for i in faces:
    path = os.path.join(dir,i)
    print(f"Loading {i}",end=' - ')
    for j in os.listdir(path):
        img = cv2.imread(os.path.join(path,j))
        img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        img = cv2.resize(img,(220,220)).flatten()
        
        x.append(img)
        y.append(faces.index(i))
    print("Success")
```

```python
import numpy as np

x = np.array(x)
y = np.array(y)
```

```python
from sklearn.svm import SVC

svc = SVC()

svc.fit(x,y)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann = Sequential()

ann.add(Dense(4,activation='relu'))
ann.add(Dense(8,activation='relu'))
ann.add(Dense(16,activation='relu'))
ann.add(Dense(16,activation='relu'))
ann.add(Dense(5,activation='softmax'))
```

```python
ann.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
x.shape,y.shape
```

```python
x = x/255
ann.fit(x,y,epochs=20,batch_size=20)
```

```python
pred_video = cv2.VideoCapture(0)

while True:
    succ, pred_image = pred_video.read()

    gray = cv2.cvtColor(pred_image,cv2.COLOR_BGR2GRAY)

    face = model.detectMultiScale(gray,minNeighbors=5)

    for (x,y,w,h) in face:

        gray_resized = cv2.resize(gray[y:y+h,x:x+w],(220,220)).flatten()

        prediction = ann.predict((gray_resized.reshape(1,-1))/255)
        
        name = faces[np.argmax(prediction)]
        cv2.rectangle(pred_image,(x,y),(x+w,y+h),(0,0,255),2)
        cv2.putText(pred_image,f"{name.upper()}",(x,y-15),cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0),2)

    cv2.imshow("Face",pred_image)
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break

pred_video.release()
cv2.destroyAllWindows()
```

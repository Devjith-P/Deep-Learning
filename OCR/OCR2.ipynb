```python
import cv2
import pytesseract
```

```python
img=cv2.imread('ocr3.jpeg')
cv2.imshow("OCR",img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
text=pytesseract.image_to_string(img)
print(text)
```

```python
dic=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
dic
```

```python
n=len(dic['text'])
n
```

```python
#dictionary
dic1={'name':['vipin','vinay','anu'],'roll':[101.102],'marks':[45,56]}
print(dic1['name'])
print(dic1['name'][0])
print(dic1[])
```

```python
for i in range(n):
    if dic['conf'][i]>=50:
        x=dic['left'][i]
        y=dic['top'][i]
        w=dic['width'][i]
        h=dic['height'][i]
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255,3))
cv2.imshow('Image',img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

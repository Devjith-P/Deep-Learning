```python
import cv2,pytesseract
```

```python
img = cv2.imread('/home/devjith/Luminar/Deep Learning/files/OCR/ocr3.jpeg')
```

```python
props = pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
```

```python
for i in range(len(props['text'])):
    if props['conf'][i] > 50:
        x,y = props['left'][i],props['top'][i]
        w,h = props['width'][i],props['height'][i]

        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

cv2.imshow("text",img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

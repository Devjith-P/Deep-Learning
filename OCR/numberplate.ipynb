```python
import cv2,pytesseract,re
```

```python
img = cv2.imread('/home/devjith/Luminar/Deep Learning/files/OCR/number_plate4.jpg')
```

```python
data = pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
```

```python
data
```

```python
text = pytesseract.image_to_string(img)
text
```

```python
full_text = ' '.join(i for i in data['text'])
for i in range(len(data['text'])):
    if re.search(r'([A-Z]{2}\s*\d{1,2}\s*[A-Z]{1,3}\s*\d{1,4}|\d{2}[A-Z]{2}\d{4})', data['text'][i]):
        print(f"Detected Number Plate is {data['text'][i]}")   
        x = data['left'][i]
        y = data['top'][i]
        w = data['width'][i]
        h = data['height'][i]

        cv2.rectangle(img, (x, y), (x+w, y+h), (0,0,255), 2)

cv2.imshow("nb", img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

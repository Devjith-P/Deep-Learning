```python
import cv2
import pytesseract
```

```python
img=cv2.imread("ocr2.jpeg")
cv2.imshow("invoice",img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
text=pytesseract.image_to_string(img)
print(text)
```

```python
dic=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
dic
```

```python
n=len(dic['text'])
```

```python
import re
```

```python
#date format in invoice with this we create a regular expression 
#here format is MM|DD|YYYY
#normal data have patterns 
#normal date pattern : YYYY|MM|DD 
for i in range(n):
    if re.search(r'(\d{1,2}/\d{1,2}/\d{4})',dic['text'][i]):
        print (dic['text'][i])
        x=dic['left'][i]
        y=dic['top'][i]
        w=dic['width'][i]
        h=dic['height'][i]
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255,3))
cv2.imshow('Date in Invoice',img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
import mediapipe as mp
import cv2
```

```python
mp_pose=mp.solutions.pose
mp_drawing=mp.solutions.drawing_utils
```

```python
pose=mp_pose.Pose()
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=pose.process(img1)
    print(result.pose_landmarks)
    if result.pose_landmarks:
    
        mp_drawing.draw_landmarks(img,result.pose_landmarks,mp_pose.POSE_CONNECTIONS)#,mp_drawing.DrawingSpec(color=(255,0,0),thickness=5,circle_radius=3),mp_drawing.DrawingSpec(color=(0,0,255),thickness=5))


    cv2.imshow('Pose',img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```

```python
import mediapipe as mp
import cv2 
```

```python
mp_detection=mp.solutions.face_detection
mp_drawing=mp.solutions.drawing_utils
```

```python
face_detect=mp_detection.FaceDetection()
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=face_detect.process(img1)
    print(result.detections)
    if result.detections:
        for x in result.detections:
            # mp_drawing.draw_landmarks(img,x)    # to form the land mark
             mp_drawing.draw_detection(img,x)

    cv2.imshow("Face",img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```

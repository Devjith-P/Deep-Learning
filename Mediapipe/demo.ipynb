```python
import mediapipe as mp
import cv2
```

```python
mphands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
```

```python
data = mphands.Hands()
```

```python
video = cv2.VideoCapture(0)

while True:
    suc, img = video.read()

    img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

    result = data.process(img1)

    print(result.multi_hand_landmarks)
    if result.multi_hand_landmarks:
        for x in result.multi_hand_landmarks:
            mp_drawing.draw_landmarks(img,x,mphands.HAND_CONNECTIONS,mp_drawing.DrawingSpec(color=(255,0,0),thickness=2,circle_radius=1),mp_drawing.DrawingSpec(color=(0,255,0),thickness=1))

    cv2.imshow("Video",img)

    if cv2.waitKey(1) & 0XFF == ord('q'):
        break


video.release()
cv2.destroyAllWindows()
```
